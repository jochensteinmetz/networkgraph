```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
#ax.plot([1.0,1.5], [5,5], color='r', linestyle='-')
# see https://stackoverflow.com/questions/19130991/matplotlib-how-to-plot-concentric-circles-at-a-given-set-of-radii
ax.plot(np.linspace(0, 1*np.pi, 100), np.ones(100)*3, color='r', linestyle='-')
ax.plot(np.linspace(0, 1*np.pi, 100), np.ones(100)*7, color='b', linestyle='-')
plt.show()
```
